```python
# Programa 1: Calcular la media y desviación estandar de una serie de datos.
```

**Datos para calcular:** DATOS_EMC_INTGRA.png

```python
import numpy as np
import pandas as pd
```

```python
Datos_EPS = [160, 591, 114, 229, 230, 270, 128, 1657, 624, 1503] #EPS es el nombre establecido para identificar a Estimated Proxy Size
Datos_DH = [15.0, 69.9, 6.5, 22.4, 28.4, 65.9, 19.4, 198.7, 38.8, 138.2] #DH es el nombre establecido para identificar a Development Hours
```

**Fórmula para calcular la media:** FORMULA_MEDIA.png


```python
# Calcular la media de Estimated Proxy Size

EPS_media = sum(Datos_EPS) / len(Datos_EPS)
print(f"Media de Estimated Proxy Size: {EPS_media}")
```

```python
# Calcular la media de Development Hours

DH_media = sum(Datos_DH) / len(Datos_DH)
print(f"Media de Development Hours: {DH_media}")
```

**Fórmula para calcular la desviación estándar**: FORMULA_DESVIACIONESTANDAR.png

```python
sumatorio_cuadrados1 = 0
for Datos1 in Datos_EPS:
  sumatorio_cuadrados1 += (Datos1 - EPS_media) ** 2

DesviacionEstandar_EPS = (sumatorio_cuadrados1/(len(Datos_EPS) -1)) ** 0.5
print(f"Desviacion estandar de Estimated Proxy Size: {DesviacionEstandar_EPS}")
```

```python
sumatorio_cuadrados2 = 0
for Datos2 in Datos_DH:
  sumatorio_cuadrados2 += (Datos2 - DH_media) ** 2

DesviacionEstandar_DH = (sumatorio_cuadrados2/(len(Datos_DH) -1)) ** 0.5
print(f"Desviacion estandar de Estimated Proxy Size: {DesviacionEstandar_DH}")
```

```python
# Resultados

pd.DataFrame([[EPS_media, DesviacionEstandar_EPS],
              [DH_media, DesviacionEstandar_DH]],
             index=[['Estimated Proxy Size', 'Development Hours']],
             columns=['Media', 'Desviación Estandar']).round(2)
```
